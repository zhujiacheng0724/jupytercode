```python
import numpy as np 
import matplotlib.pyplot as plt
import skimage
```

```python
import skimage.draw as draw
img= np.zeros([100,100])
h,w=draw.line(1,1,95,95)
img[h,w]=1
plt.imshow(img)
```

```python
import skimage.transform as transform
#霍夫变化-线检测

h,theta,d=transform.hough_line(img)

plt.imshow(np.log(1+h))
```

```python
#霍夫变化-圆检测
img= np.zeros([200,200])
h,w=draw.circle(150,50,30)
img[h,w]=1
#画一个矩形
img[40:60,60:80]=1

plt.imshow(img)
```

```python
radii =np.arange(5,50,5)
print(radii)

circle=transform.hough_circle(img,radii)

plt.figure("circle",figsize=(10,10))
plt.subplot(121)
plt.imshow(circle[3,:,:])
print(np.max(circle[3,:,:]))
plt.subplot(122)
plt.imshow(circle[5,:,:])
print(np.max(circle[5,:,:]))
```

```python
#边界检测
import skimage.measure as measure

res=measure.find_contours(img,0.5)

print(len(res))
```
