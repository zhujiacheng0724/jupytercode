```python
import matplotlib.pyplot as plt 
import numpy
import skimage
```

```python
import skimage.io as io
```

```python
img =io.imread("zjcdy.JPG")
plt.imshow(img)
```

```python
img_blackwhite= skimage.color.rgb2gray(img)

plt.imshow(img_blackwhite,plt.cm.gray)
```

```python
import skimage.filters as filters
img_e= filters.gaussian(img,sigma=20)

plt.imshow(img_e)
```

```python
import skimage.feature as features
img_f =filters.roberts(img_blackwhite)

plt.imshow(img_f)
```

```python
img_f1=filters.prewitt_h(img_blackwhite)
img_f2=filters.prewitt_v(img_blackwhite)
plt.figure("prewitt",figsize=(8,8))
plt.subplot(121)
plt.imshow(img_f1,plt.cm.gray)
plt.subplot(122)
plt.imshow(img_f2,plt.cm.gray)

plt.show()
```

```python
img_g=filters.prewitt(img_blackwhite)
plt.imshow(img_g)
```

```python
img_g=filters.sobel(img_blackwhite)
plt.imshow(img_g)
```

```python
img_g=features.canny(img_blackwhite,sigma=1)
plt.imshow(img_g)
```

```python
import skimage.transform as transform
img_n=transform.resize(img,[100,80])
plt.imshow(img_n)
```

```python
img_gray=skimage.color.rgb2gray(img_n)
img_gray[img_gray<0.5]=1
img_gray[img_gray>0.5]=0

plt.imshow(img_gray)
```
